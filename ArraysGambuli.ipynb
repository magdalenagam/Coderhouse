```python
#factorización de un número

import numpy as np

random2=np.random.randint(1,11, size=(1,5))
random_num=random2.min()

def factorial(random_num):
  fact=1
  for n in range(1,random_num+1):
    fact*=n
  print(random2)
  print("Número random:",random_num)
  print("El factorial del número es:",fact)

factorial(random_num)
```

```python
#suma de serie de números

import numpy as np

random3=np.random.randint(0,25, size=(3,3))
min=random3.min()
max=random3.max()

def suma_serie(min,max):
  sum=0
  if min<max:
    for i in range(min,max+1):    
      sum += i 
    for i in range(min,max+1): 
      if i == max:
        print(i,end = '')
      else: print(i,"+ ",end = '')
    print(" =",sum)
  else:
    print("Operación inválida. El primer número de la serie debe ser menor al segundo.")

print("Array random:\n",random3)
print("Número mínimo:",min)
print("Número máximo:",max)
suma_serie(min,max)
```
