```python
#Objeto de series de tiempo con indice igual al año
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

medals_summer = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Documents/summer_olympic_medals.csv', sep=',', encoding='latin-1', header=[0])
medals_summer = medals_summer.fillna(0);
medals_summer['Year'] = medals_summer['Year'].astype(int)
medals_summer = medals_summer.set_index('Year')

medals_summer
```

```python
#Separar nombres de apellidos

medals_summer['Athlete'].str.split(',', n=1, expand=True)
```

```python
#Medidas resumen del conj de datos
print(medals_summer.info())
print('Medallas ganadas en cada deporte por país:\n', medals_summer.groupby(['Country','Sport']).size())
print('Género de los participantes:\n', medals_summer.groupby(['Year', 'Gender']).size())
```

```python
#Que pais ganó más medallas?
most_medals_country = medals_summer['Country'].value_counts().idxmax()
print('El país con más medallas es: ',most_medals_country)
```

```python
#Cuantas medallas ganaron los hombres cada año

medals_men = medals_summer.query('Gender == "Men"')
medals_men.groupby(['Year', 'Gender']).size()
```
