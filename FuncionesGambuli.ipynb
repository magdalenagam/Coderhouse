```python
#CALCULADORA DE FACTORIZACIÓN

def factorial(num):
  fact=1
  if num<0:           
    print("Solo se puede sacar el factorial de un número positivo")
  elif num==0:        
    print("El factorial de 0 es 1")
  else:
    for n in range(1,num+1):
      fact*=n
  print("El factorial de ",num," es ",fact)

num=int(input("Ingrese número: "))
factorial(num)
 

   
```

```python
#SUMA DE SERIE DE NÚMEROS

def suma_serie(x,y):
  sum = 0
  if x<y:
    for i in range(x,y+1):     #loop de suma
      sum += i 
    for i in range(x,y+1):     #loop de impresión
      if i == y:
        print(i,end = '')
      else: print(i,"+ ",end = '')
    print(" =",sum)
  else:
    print("Operación inválida. El primer número de la serie debe ser menor al segundo.")

x = int(input("Ingrese el primer número: "))
y = int(input("Ingrese el segundo número: "))
suma_serie(x,y)
```
